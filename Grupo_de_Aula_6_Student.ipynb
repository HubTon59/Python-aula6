````python
class Livro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.categoria = ['Romance','Aventura','ficção','Terror']

class Livraria:
    def __init__(self):
        self.livros = {}
        
    #-mostrar livros disponives
        
class Sistema_gestão(Livraria):
    def __init__(self):
        pass
    
    def novo_livro(self):
        self.titulo = input("Digite o título do livro: ")
        self.autor = input("Digite o autor do livro: ")
        self.categoria = self.categoria
        
        print("---CATEGORIAS---")
        i = 0
        for item in self.categoria:
            i += 1
            print(i,'-', self.categoria)
        
        self.index = input('Digite o numero de acordo com a categoria do livro: ')

        novo_livro = Livro({'Título': self.titulo, 'Autor': self.autor, 'Categoria': self.categoria})
        self.livros.append(novo_livro)
        return print(f'O livro {self.titulo} foi adicionado com sucesso!!!!!')
    
    def adicionar_categoria(self, categoria):
        nova_categoria = input('Digite o nome da nova categoria: ')
        categoria.append(nova_categoria)
        return f'Categoria {nova_categoria} adicinada com sucesso!!'

    def listar_livros(self):
        print("Livros disponíveis: ")
        
        for item in self.livros:
            print(self.livros['Título'],'\n')
            print(self.livros['Autor'],'\n')
            print(self.livros['Categoria'],'\n','\n')
        
        return 

        


class Calogo(Livraria):
    def __init__(self):
        pass
    
    def verificacao_categoria(self):
        #list_romance = []
        #list_aventura = []
        #list_ficcao = []
        #list_terror = []
        list_outros = {}

        for i in self.livros:
            self.livros.

    
    
    
    
    
    # integrado a novo catálogos
    #-acessar o catálogo do livro

    #-criar catálogos de tópicos específicos
    #-atulizar catalogo de livros




class Livraria:
    def __init__(self):
        self.catalogo = []
        self.adicionar_livro(Livro(
            'Jantar Secreto',
            'Raphael Montes',
            'Jantar secreto conta a história de um grupo de amigos que se muda para o Rio de Janeiro e passa a organizar jantares misteriosos.',
            'Permitir que o usuário sinta o gosto de comidas'
        ))
        self.adicionar_livro(Livro(
            '1984',
            'George Orwell',
            'Mostra uma futura realidade distópica ocorrida no ano de 1984',
            'Consegue ver'
        ))
        self.adicionar_livro(Livro(
            'O Feiticeiro de Oz',
            'L. Frank Baum',
            'Um clássico conto de aventuras em um mundo mágico cheio de maravilhas e perigos.',
            'Transporta para mundos mágicos'
        ))
        self.adicionar_livro(Livro(
            'A Ilha do Tesouro',
            'Robert Louis Stevenson',
            'A emocionante história de um jovem que se aventura em busca de um tesouro escondido.',
            'Oferece mapas secretos'
        ))
 
    def adicionar_livro(self, livro):
        self.catalogo.append(livro)
        return f'O livro "{livro.titulo}" foi adicionado ao catálogo.'
 
    def listar_livros(self):
        if not self.catalogo:
            return 'O catálogo está vazio.'
 
        lista_livros = '\n\n'.join(str(livro) for livro in self.catalogo)
        largura = 50
        return f'\n{(f"Livros:").center(largura, "*")}\n{lista_livros}'
 
    def listar_livros_emprestados(self):
        emprestados = [livro for livro in self.catalogo if livro.emprestado]
        if not emprestados:
            return 'Nenhum livro está emprestado.'
 
        lista_emprestados = '\n\n'.join(str(livro) for livro in emprestados)
        return f'Livros Emprestados:\n{lista_emprestados}'
 
    def emprestar_livro(self, titulo):
        livro = self.encontrar_livro(titulo)
        if livro:
            return livro.emprestar()
        return f'O livro "{titulo}" não foi encontrado.'
 
    def devolver_livro(self, titulo):
        livro = self.encontrar_livro(titulo)
        if livro:
            return livro.devolver()
        return f'O livro "{titulo}" não foi encontrado.'
 
    def encontrar_livro(self, titulo):
        titulo_lower = titulo.lower()
        for livro in self.catalogo:
            if livro.titulo.lower() == titulo_lower:
                return livro
        return None
 
    def listar_poderes_magicos(self):
        poderes = set(livro.poderes_magicos for livro in self.catalogo)
        if not poderes:
            return 'Nenhum poder mágico disponível.'
 
        return 'Poderes Mágicos Disponíveis:\n' + '\n'.join(f'- {poder}' for poder in poderes)
 
    def buscar_por_poder_magico(self, poder_magico):
        poder_magico_lower = poder_magico.lower()
        livros_encontrados = [livro for livro in self.catalogo if livro.poderes_magicos.lower() == poder_magico_lower]
        if not livros_encontrados:
            return f'Nenhum livro encontrado com o poder mágico "{poder_magico}".'
 
        return '\n\n'.join(str(livro) for livro in livros_encontrados)
    
    def emprestar(self):
        if self.emprestado:
            return f'O livro "{self.titulo}" já está emprestado.'
        self.emprestado = True
        self.data_emprestimo = datetime.now()
        self.data_devolucao = self.data_emprestimo + timedelta(days=20)  # Devolução em 20 dias
        return (f'O livro "{self.titulo}" foi emprestado com sucesso em {self.data_emprestimo}. '
                f'A devolução está prevista para {self.data_devolucao}.')
 
    def devolver(self):
        if not self.emprestado:
            return f'O livro "{self.titulo}" não está emprestado.'
        self.emprestado = False
        self.data_emprestimo = None
        self.data_devolucao = None
        return f'O livro "{self.titulo}" foi devolvido.'
 
    def __str__(self):
        status = 'Emprestado' if self.emprestado else 'Disponível'
        emprestimo = (f' - Desde: {self.data_emprestimo} '
                      f'(Devolução prevista: {self.data_devolucao})') if self.data_emprestimo else ''
        return (f'Título: {self.titulo}\n'
                f'Autor: {self.autor}\n'
                f'Descrição: {self.descricao}\n'
                f'Poderes Mágicos: {self.poderes_magicos}\n'
                f'Status: {status}{emprestimo}')
````
